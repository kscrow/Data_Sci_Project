```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
heartdata = pd.read_csv('heart_kaggle_project.csv')
heartdata.head()
```

```python
heartdata = heartdata.reindex(np.random.permutation(heartdata.index))
heartdata.head(10)
```

```python
heartdata = heartdata.reset_index(drop=True)
heartdata.head(10)
```

```python
heartdata[['age', 'trestbps', 'chol', 'thalach']] = StandardScaler().fit_transform(heartdata[['age', 'trestbps', 'chol', 'thalach']])
heartdata.head(10)
```

```python
heartdata.describe()
```

```python
y = heartdata['target']
y.head()
```

```python
x = heartdata.drop('target', axis = 1)
x.head()
```

```python
#Random Forest here
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2) 
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=10)
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)

from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
